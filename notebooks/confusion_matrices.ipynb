```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('azizsambo58@gmail.com', 'baseline')
fig, ax = plot_confusion_matrix(
    conf_mat=np.array([
        [1400, 297],
        [571, 162],
    ]),
    show_absolute=True,
    show_normed=False,
    colorbar=True,
    class_names = ['no stress', 'stress'],
    figsize=(6, 6)
)
plt.show()
```
